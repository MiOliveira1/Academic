```python
median_PGC = df['PGC'].median()
print(median_PGC)

median_DBP = df['DBP'].median()
print(median_DBP)

median_TST = df['TST'].median()
print(median_TST)

median_SIC = df['SIC'].median()
print(median_SIC)

median_BMI = df['BMI'].median()
print(median_BMI)
```

```python
df.iloc[:,4:5] = df.iloc[:,4:5].replace(np.nan,125) #substituição da coluna SIC dos valores NaN pela mediana 
```
