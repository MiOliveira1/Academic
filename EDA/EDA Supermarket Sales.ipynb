# Task 2:univariate analysis 

```python
#what does the customer ranking looks like? is it skewed (enviesado)?
#gráfico com distribuição uniforme, não há nenhum rating que se destaque muito mais, tudo mais ou menos na mesma escala 

sns.displot(df['Rating'])
plt.axvline(x=np.mean(df['Rating']), c='red', ls='--', label='mean')# desenhar a media
plt.axvline(x=np.percentile(df['Rating'],25), c='green', label='percentile 25')
plt.axvline(x=np.percentile(df['Rating'],75), c='green', label ='percentile 75')
plt.legend() #imprime legenda
```

```python
df.hist(figsize=(20,20)) #figsize aumenta o tamanho - este gráfico não é aconselhável para grandes datasets 

#margem é sempre igual 
#tax right skewed 
#total e gross incrome relacionados 
```
